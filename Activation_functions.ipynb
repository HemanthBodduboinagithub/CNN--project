```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
def sigmoid(x):
  num = 1
  den = 1 + math.e**(-x)
  return num/den
```

```python
sigmoid(4)
```

```python
x = np.arange(-10, 10)
y = sigmoid(x)
plt.plot(y)
```

```python
x = np.arange(-20,20)
y = sigmoid(x)
plt.vlines(x=0, ymin=0, ymax=1, color = 'r')
plt.hlines(y=0.5, xmin=-20, xmax=20, color = 'r')
plt.plot(x,y)
```

```python
def tanh(x):
  num = math.e**(x)-math.e**(-x)
  den = math.e**(x)+math.e**(-x)
  return num/den
```

```python
x = np.arange(-20,20)
y = tanh(x)
plt.vlines(x=0, ymax=1, ymin=-1, color = 'r')
plt.hlines(y=0, xmax=20, xmin=-20, color = 'r')
plt.plot(x,y)
```
